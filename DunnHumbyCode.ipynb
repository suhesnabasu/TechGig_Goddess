```
def returnDf(individual_customer):
  global products_table, relevancy_output
  individual_customer.drop_duplicates(subset=['product'], inplace=True)
  products = individual_customer['product'].tolist()
  #individual_customer.head()
  count_product = 0
  product = []
  customer = []
  relevancy = []
  for item in products:
    #print("In loop")
    #print(int(products_table.loc[item]))
    if count_product == 8:
      break

    if int(products_table.loc[item]) != 0:
      #print("Checking Volume of " + str(item))

      if item in list(exclusion_table.index):
        #print("Item Present")
        #print(type(exclusion_table.loc[item]['product2'].tolist()))
        x = exclusion_table.loc[item]['product2'].tolist()
        if isinstance(x, int):
          exclusion_list = [x]
        else:
          exclusion_list = x
        #exclusion_list = [exclusion_table.loc[item]['product2'].tolist()]
        #print(exclusion_list)
        flag = 0
        for element in exclusion_list:
          if element in product:
            #print(element)
            #print("Element in product assigned")
            flag = 1
            break
        
        if flag != 1:
          product.append(item)
          products_table.loc[item] -= 1
          count_product += 1
          score = individual_customer[individual_customer['product'] == item].values[0][1]
          #print("printing score")
          #print(score)
          relevancy.append(score)
          cust = str(individual_customer[individual_customer['product'] == item].index[0])
          print("Processing Customer "+ str(cust))
          customer.append(cust)

      else:
        #print("Item Absent")
        product.append(item)
        products_table.loc[item] -= 1
        count_product += 1
        score = individual_customer[individual_customer['product'] == item].values[0][1]
        #print("printing score")
        #print(score)
        relevancy.append(score)
        cust = str(individual_customer[individual_customer['product'] == item].index[0])
        customer.append(cust)
  
  temp = pd.DataFrame({'customers': customer, 'product': product, 'relevancy_score': relevancy})
  temp.set_index('customers', inplace=True)
  frames = [relevancy_output, temp]
  relevancy_output = pd.concat(frames)
  print("Done Processing Customer ")
  #relevancy_output.head(10)
  #print(temp.head(10))  

```

```
 for values in relevacy_table.index.unique().tolist():
   returnDf(relevacy_table.loc[values].copy())
```
